```python
import PyPDF2
import io
import pandas as pd
```

```python
# creating a pdf file object 
pdfFileObj = open('state_reports_latest.pdf', 'rb') 
```

```python
# creating a pdf reader object 
pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
```

```python
p25=pdfReader.getPage(25)
p25.extractText().splitlines()
```

```python
dictsites = {}
dict100sites={}
for pn in range(25,49):
    pg=pdfReader.getPage(pn)
    lines=pg.extractText().splitlines()
#    print(lines)
    idx=0
    for l in lines:
#        print('idx='+str(idx)+ ' ' + l)
        if "%" in l:
            if(not lines[idx-1].replace(',','').isnumeric()):
                dictsites[lines[idx-1]]=int(lines[idx+1].replace(',',''))
            elif (not lines[idx-2].replace(',','').isnumeric()):
                dictsites[lines[idx-2]]=int(lines[idx+1].replace(',',''))
            elif (not lines[idx-3].replace(',','').isnumeric()):
                dictsites[lines[idx-3]]=int(lines[idx+1].replace(',',''))
            elif (not lines[idx-4].replace(',','').isnumeric()):
                dictsites[lines[idx-4]]=int(lines[idx+1].replace(',',''))
            candidate=l.replace('%','')
            if(candidate.isnumeric()):
                if int(candidate)>99:
                    if(not lines[idx-1].replace(',','').isnumeric()):
                        dict100sites[lines[idx-1]]=int(lines[idx+1].replace(',',''))
                    elif (not lines[idx-2].replace(',','').isnumeric()):
                        dict100sites[lines[idx-2]]=int(lines[idx+1].replace(',',''))
                    elif (not lines[idx-3].replace(',','').isnumeric()):
                        dict100sites[lines[idx-3]]=int(lines[idx+1].replace(',',''))
                    elif (not lines[idx-4].replace(',','').isnumeric()):
                        dict100sites[lines[idx-4]]=int(lines[idx+1].replace(',',''))

                    

#        if "100%" in l:
#            dict100sites[lines[idx-4]]=int(lines[idx+1].replace(',',''))
#            print(lines[idx-4] + ' = '+ str(int(lines[idx+1].replace(',',''))))
        idx=idx+1
        
        
```
